```python
AreaDict  = {'Durham, NC':'RTP', 
                'Wake, NC':'RTP', 
                'Alameda, CA':'Bay Area', 
                'Contra Costa, CA':'Bay Area', 
                'Marin, CA':'Bay Area', 
                'Napa, CA':'Bay Area', 
                'San Francisco, CA':'Bay Area', 
                'San Mateo, CA':'Bay Area',
                'Santa Clara, CA':'Bay Area', 
                'Solano, CA':'Bay Area', 
                'Sonoma, CA':'Bay Area', 
                'Bronx, NY':'NYC', 
                'Kings, NY':'NYC',
                'New York, NY' : 'NYC',
                'Queens, NY':'NYC',
                'Richmond, NY':'NYC',
                'Middlesex, MA': 'Boston',
                'Norfolk, MA' : 'Boston',
                'Suffolk, MA' : 'Boston',
                'United States *' : 'USA'}

df_industry['Region'] = pd.Series(AreaDict, name='GeoName')
df_tech['Region'] = pd.Series(AreaDict, name='GeoName')
df_capita['Region'] = pd.Series(AreaDict, name='GeoName')
```

```python
industry_mean = df_industry.groupby("Region").mean()
tech_mean = df_tech.groupby("Region").mean()
industry_mean = industry_mean.div(1000000)
tech_mean = tech_mean.div(1000000)
capita_sum = df_capita.groupby("Region").sum()
```

```python
NC_top10_tech = NC_tech.nlargest(n=10, columns=['2020'])
NC_top10_tech = NC_top10_tech.T
NC_top10_tech = NC_top10_tech.div(1000000)
```

```python
industry_mean_rotate = industry_mean.T
industry_mean_rotate_noUSA = industry_mean_rotate.drop(columns="USA")
```

```python
tech_mean_rotate = tech_mean.T
tech_mean_rotate_noUSA = tech_mean_rotate.drop(columns="USA")
```

```python
capita_sum_T = capita_sum.T
capita_sum_T = capita_sum_T.div(1000000)
capita_change = capita_sum_T.pct_change()
capita_change = capita_change*100
```

```python
tech_mean_noUSA = tech_mean.drop(index="USA")
tech_per_capita = np.divide(tech_mean_noUSA, capita_sum)
tech_per_capita = tech_per_capita*1000000
```

```python
tech_per_capita_t = tech_per_capita.T
```

The Research Triangle Park is the center of three major universities in North Carolina; Duke, UNC-Chapel Hill, and NC State. With these universities as feeders, the research triangle park became a natural area for biotech, technology companies, and professional services. Compared to the rest of North Carolina, the RTP counties (Wake and Durham) make up 2 of the top 3 GDP of professional, scientific, and technical services counties in North Carolina. They are two of the only three counties with a noticeable increase in this category over the last 20 years.

To further understand the economy of the RTP, I compared the private industry GDP of RTP to the greater Boston area, the Bay area, and NYC. Looking at both the Total Private Industry GDP and the GDP of Professional, scientific and technical services, the RTP area is well below the other three regions. However, RTP has a much smaller population compared to those regions and has had more growth over the last 20 years compared to all three other areas. When we take this into account, RTP has a higher GDP per capita in professional, scientific, and technical services than the Bay Area and NYC, and is much closer to Boston. In the future, it will be interesting to see how the GDP of RTP grows as companies, like Apple and Google, commit campuses to the area. It is an area to keep an eye on in the upcoming years.
